**Python Basic Questions**

1. What is Python, and why is it popular?
    - Python is a high-level, interpreted, general-purpose programming language known for its readability and simplicity.

      It is popular because:

      *   Simple and clean syntax
      *   Large standard library and ecosystem
      *   Cross platform compatibility
      *   Strong community support

2. What is an interpreter in Python?
    - An interpreter is a program that executes Python code line-by-line instead of compiling it into machine code beforehand.

3. What are pre-defined keywords in Python?
    - Keywords are reserved words in Python that have predefined meanings and cannot be used for other purposes.

      Examples: if, else, elif, while, for, True, False

4. Can keywords be used as variable names?
    - No, keywords cannot be used as variable names because they are reserved by Python syntax.

5. What is mutability in Python?
    - Mutability refers to whether an object’s value can be changed after creation.

6. Why are lists mutable, but tuples are immutable?
    - Lists are designed for dynamic data storage(add/remove/update elements)
    - Tuples are designed for fixed data → provide:

        Faster execution

        Memory efficiency

        Data integrity (cannot be changed accidentally)

7. What is the difference between "==" and "is" operators in Python?
    - == → Compares values

    - is → Compares memory location (identity)

8. What are logical operators in Python?
    - Logical operators are used to combine conditional statements.
      - and → True if both conditions are true
      - or → True if at least one condition is true
      - not → Reverses the condition

9. What is type casting in Python?
    - Type casting is the process of converting one data type into another.

10. What is the difference between implicit and explicit type casting?
    - Implicit Casting: Python automatically converts types

    - Explicit Casting: Programmer converts types using functions

11. What is the purpose of conditional statements in Python?
    - Conditional statements are used to control program flow based on conditions. They help in decision-making in programs.

12. How does the elif statement work?
    - elif (else-if) is used to check multiple conditions sequentially.

      1) Executes the first true condition

      2) Skips the rest once matched

13. What is the difference between for and while loops?
    - For Loop
      - Iterates over sequence
      - Iterations are fixed
      - Example: for i in range(5)

    - While Loop
      - Runs based on condition
      - Iterations are unknown
      - Example: while i < 5

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.


```python
my_list = [10, 20, 30]

my_tuple = (10, 20, 30)

my_list[0] = 100

print("Modified list:", my_list)

try:
    my_tuple[0] = 100
except TypeError:
    print("Tuple cannot be modified")

print("Tuple remains unchanged:", my_tuple)
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def demonstrate_mutability(num, lst):

    num = num + 10

    lst.append(100)

    print("Inside function:")
    print("num =", num)
    print("list =", lst)

x = 5
y = [1, 2, 3]

demonstrate_mutability(x, y)

print("\nOutside function:")
print("num =", x)
print("list =", y)
```

10. Write code to demonstrate type casting with list elements.

```python
str_list = ["1", "2", "3", "4"]

int_list = [int(i) for i in str_list]
float_list = [float(i) for i in str_list]

print("Original list:", str_list)
print("Integer list:", int_list)
print("Float list:", float_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
total = 0

for i in range(1, 51):
    if i % 2 == 0:
        total += i

print(total)
```

14. Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")

i = len(text) - 1
reverse = ""

while i >= 0:
    reverse += text[i]
    i -= 1

print(reverse)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))

fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print(fact)
```
